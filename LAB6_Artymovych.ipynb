```python


import pandas as pd

# 1. Завантажити
data = pd.read_csv("la-crime.csv")
```

```python
# 2. Перші 5 записів
print(data.head())
```

```python
# 3. Розмір таблиці
print("Розмір:", data.shape)
```

```python
# 4. Назви стовпців
print("Стовпці:")
print(data.columns)
```

```python
# 5. Типи даних
print("Типи даних:")
print(data.dtypes)
```

```python
# 6. Унікальні значення
print("Унікальні значення:")
print(data.nunique())
```

```python
# 7. Пропущені значення
print("Пропущені:")
print(data.isnull().sum())
```

```python
# АНАЛІЗ ДАНИХ (ВАРІАНТ 1)

# 1. Яке походження найчастіше є жертвами?
print(data["Vict Descent"].value_counts())
```

```python
# 2. Чи вірно, що жінки частіше стають жертвами?
print(data["Vict Sex"].value_counts())
```

```python
# 3. Розподіл кількості злочинів за віком (гістограма)
import matplotlib.pyplot as plt

data["Vict Age"].hist(bins=30)
plt.title("Розподіл злочинів за віком")
plt.xlabel("Вік")
plt.ylabel("Кількість")
plt.show()
```

```python
# 4. Як відрізняється ймовірність для чоловіків і жінок залежно від віку
import seaborn as sns

sns.histplot(data=data, x="Vict Age", hue="Vict Sex", bins=30, kde=True)
plt.title("Розподіл віку жертв за статтю")
plt.show()
```

```python
# 5. В якому віковому проміжку чоловіки частіше стають жертвами
male = data[data["Vict Sex"] == "M"]

male["Vict Age"].hist(bins=30)
plt.title("Розподіл віку чоловіків-жертв")
plt.xlabel("Вік")
plt.ylabel("Кількість")
plt.show()
```

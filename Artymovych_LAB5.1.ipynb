```python
def find_primes(n):
    if not isinstance(n, int):
        raise TypeError("Потрібно ввести ціле число.")

    if n < 2:
        return []

    primes = []

    for num in range(2, n + 1):
        is_prime = True

        for i in range(2, int(num ** 0.5) + 1):
            if num % i == 0:
                is_prime = False
                break

        if is_prime:
            primes.append(num)

    return primes
```

```python
print("Прості числа до 10:", find_primes(10))
print("Прості числа до 20:", find_primes(20))
print("Прості числа до 50:", find_primes(50))
```

```python
try:
    print(find_primes("10"))
except Exception as e:
    print("Помилка:", e)

try:
    print(find_primes(5.5))
except Exception as e:
    print("Помилка:", e)

print("Для N = 1:", find_primes(1))
```

```python
import matplotlib.pyplot as plt

values_n = [10, 20, 30, 40, 50]
counts = [len(find_primes(n)) for n in values_n]

plt.figure(figsize=(8, 5))
plt.bar(values_n, counts)
plt.title("Кількість простих чисел до N")
plt.xlabel("N")
plt.ylabel("Кількість простих чисел")
plt.grid(True)
plt.show()
```
